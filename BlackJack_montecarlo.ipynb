```python
## Monte Carlo method ##
def MC(pi,gamma=0.9,itrnum =10000,Q=np.zeros([528,2])):
    V =  np.zeros([22,12,2])
    
    #Q =  np.zeros([528,2])
    ReturnsCounter = np.zeros([528,2])
    Returns = [[[] for j in range(2) ] for i in range(528)]
    counter = 0
    for j in tqdm(range(itrnum)):
        epi = []
        deck = deckinit()
        C_deck = []
        U_deck = []
        C_deck = hit(C_deck,deck)
        U_deck = deal(deck)
        s0= total(U_deck);
        s1= C_deck[0];
        if aceCheck(U_deck):
            s2 = 1
        else:
            s2 = 0
        
        # 초기상태 저장
        A = pi[s0][s1][s2]
        R = 0
        S = stoS(s0,s1,s2)
        epi_=[S,A,R]
        epi.append(epi_)
        # 다음 pi 에 의해 action 결정
        while(A == 1):
            U_deck = hit(U_deck,deck)
            s0 = total(U_deck);
            if s0 > 21:
                #bust
                R = -10
                epi_=[0,0,R]
                epi.append(epi_)
                break
            if s0 == 21:
                #black jack
                R = 0
                break    
            #s1 = s1, s2 = s2
            if aceCheck(U_deck):
                s2 = 1
            # 다음 pi 에 의해 action 결정
            A = pi[s0][s1][s2]
            #약간의 보상 증정, 어쨋든 21에 가까워졌으므로
            R = (s0 - 4)/4
            S = stoS(s0,s1,s2)
            epi_=[S,A,R]
            epi.append(epi_)
        if R != -10:
            while total(C_deck) < 16:
                C_deck = hit(C_deck,deck)
            
        R = score(C_deck, U_deck)
        epi_=[0,0,R]
        epi.append(epi_)
        # 상태 Seq 완성
        
        G = 0 
      #  print(epi)
        for i in range(len(epi)-2,-1,-1):
            G = gamma * G + epi[i+1][2]
            S = epi[i][0]
            A = epi[i][1]
            
            ReturnsCounter[S][A] = ReturnsCounter[S][A] + 1
            if ReturnsCounter[S][A]>=2:
                r = ReturnsCounter[S][A]
                Returns[S][A] = (r-1) * Returns[S][A] / r +  G/ r  
            else : 
                Returns[S][A] = G
     #       Returns[S][A].append(G)
            Q[S,A] = Returns[S][A]
          # Q[S,A] = np.mean(Returns[S][A])
        
        #정책 업데이트
        for i in range(len(epi)-2,-1,-1):
            S = epi[i][0]
            if Q[S][0] < Q[S][1]:
                (s0,s1,s2) = Stos(S)
                pi[s0][s1][s2] = 1
            else:
                (s0,s1,s2) = Stos(S)
                pi[s0][s1][s2] = 0
                
        counter+=1
    return Q
```
